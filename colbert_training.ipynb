```python

def generate_examples_triples(filepath, collection_path=None, queries_path=None):
    with open(filepath, encoding="utf-8") as f:
        for (idx, line) in enumerate(f):
            query_id, pos_id, neg_id = line.rstrip().split("\t")
            yield int(query_id), int(pos_id), int(neg_id)
```

```python
def count_triples(filepath):
    with open(filepath, encoding="utf-8") as f:
        for (idx, line) in enumerate(f):
          num=idx

    return num
```

```python
triples = []
gen = generate_examples_triples(triplet)
for i in range(0,10000000):
  triple = next(iter(gen))
  print(triple)
  triples.append(triple)

triples
```
